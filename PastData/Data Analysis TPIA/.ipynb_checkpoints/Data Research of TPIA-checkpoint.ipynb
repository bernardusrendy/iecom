```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame as df
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import r2_score
```

```python
data=pd.read_excel('./TPIA.xlsx')
```

```python
data.head()
```

```python
plt.plot(data['close'])
```

Non linearity of the model is observed at t=420, therefore we start deriving the model at t=420 forward

```python
series=data['close'][420:]
```

```python
result = seasonal_decompose(series, model='additive', freq=20,two_sided=False)
result.plot()
plt.show()
```

If we analyze every structure of the decomposition, we can easily see that the residual part consisted of a sinusoidal average 

```python
residualfit=result.resid[20:]
```

```python
((np.max(residualfit)-np.min(residualfit))/2)+np.min(residualfit)
```

```python
residualfit.shape
```

```python
t=np.linspace(0,75,76)
```

```python
def sinusoidal(t, coeffs):
    return ((coeffs[0] * np.sin(coeffs[1] * t + coeffs[2]) + coeffs[3]))
lsm_sinusoidal=sinusoidal(t,[700,0.16,-0.3,135.573])
```

```python
plt.plot(lsm_sinusoidal)
plt.scatter(t, residualfit, label='Data')
```

```python
series.shape
```

```python
residualfit=result.resid[23:]
```

```python
x0 = np.array([700,0.15,1,135.573], dtype=float)
def test_func(t, coeffs):
    return ((coeffs[0] * np.sin(coeffs[1] * t + coeffs[2]) + coeffs[3]))
def residuals(coeffs, y, t):
    return (y - test_func(t, coeffs))
from scipy.optimize import leastsq
t=np.linspace(0,72,73)
x, flag = leastsq(residuals, x0, args=(residualfit, t))
x
```

```python
plt.scatter(t, residualfit, label='Data')
plt.plot(t, test_func(t,x),
         label='Fitted function')

plt.legend(loc='best')

plt.show()
```
